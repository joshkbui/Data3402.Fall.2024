```python
import matplotlib.pyplot as plt
import numpy as np

def create_pair_plot(data, feature_indices, title):
    num_features = len(feature_indices)

    # Create a grid of subplots
    fig, axs = plt.subplots(num_features, num_features, figsize=(12, 12))
    plt.suptitle(title, fontsize=16)

    for i in range(num_features):
        for j in range(num_features):
            if i == j:
                # Histogram on the diagonal
                axs[i, j].hist(data[:, feature_indices[i]], bins=30, color='gray', alpha=0.7)
                axs[i, j].set_xlabel(f'Feature {feature_indices[i]}')
                axs[i, j].set_ylabel('Count')
            else:
                # Scatter plot on the off-diagonal
                axs[i, j].scatter(data[:, feature_indices[j]], data[:, feature_indices[i]], alpha=0.5, s=10)
                axs[i, j].set_xlabel(f'Feature {feature_indices[j]}')
                axs[i, j].set_ylabel(f'Feature {feature_indices[i]}')

    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout
    plt.show()

# Example usage
# Assuming `data` is your SUSY sample with low and high-level features.
low_level_indices = [0, 1, 2, 3]  # Replace with actual indices
high_level_indices = [4, 5, 6, 7]  # Replace with actual indices

create_pair_plot(data, low_level_indices, "Low-Level Feature Pair Plot")
create_pair_plot(data, high_level_indices, "High-Level Feature Pair Plot")
```

```python
def create_pair_plot_optimized(data, feature_indices, title):
    num_features = len(feature_indices)

    # Precompute histograms for the diagonal
    hist_data = {}
    for idx in feature_indices:
        hist_data[idx] = np.histogram(data[:, idx], bins=30)

    # Create a grid of subplots
    fig, axs = plt.subplots(num_features, num_features, figsize=(12, 12))
    plt.suptitle(title, fontsize=16)

    for i in range(num_features):
        for j in range(num_features):
            if i == j:
                # Use precomputed histogram data
                bin_edges = hist_data[feature_indices[i]][1]
                counts = hist_data[feature_indices[i]][0]
                axs[i, j].bar(bin_edges[:-1], counts, width=np.diff(bin_edges), color='gray', alpha=0.7, align='edge')
                axs[i, j].set_xlabel(f'Feature {feature_indices[i]}')
                axs[i, j].set_ylabel('Count')
            else:
                # Scatter plot on the off-diagonal
                axs[i, j].scatter(data[:, feature_indices[j]], data[:, feature_indices[i]], alpha=0.5, s=10)
                axs[i, j].set_xlabel(f'Feature {feature_indices[j]}')
                axs[i, j].set_ylabel(f'Feature {feature_indices[i]}')

    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout
    plt.show()

# Example usage for optimized version
create_pair_plot_optimized(data, low_level_indices, "Optimized Low-Level Feature Pair Plot")
create_pair_plot_optimized(data, high_level_indices, "Optimized High-Level Feature Pair Plot")
```

```python
import numpy as np
from tabulate import tabulate

def compute_matrices(data, low_level_indices, high_level_indices):
    # Compute covariance matrix for all features
    covariance_matrix_all = np.cov(data, rowvar=False)
    correlation_matrix_all = np.corrcoef(data, rowvar=False)

    # Compute covariance and correlation for low-level features
    covariance_matrix_low = np.cov(data[:, low_level_indices], rowvar=False)
    correlation_matrix_low = np.corrcoef(data[:, low_level_indices], rowvar=False)

    # Compute covariance and correlation for high-level features
    covariance_matrix_high = np.cov(data[:, high_level_indices], rowvar=False)
    correlation_matrix_high = np.corrcoef(data[:, high_level_indices], rowvar=False)

    return (covariance_matrix_all, correlation_matrix_all,
            covariance_matrix_low, correlation_matrix_low,
            covariance_matrix_high, correlation_matrix_high)

def create_table(covariance, correlation, feature_names):
    # Create headers
    header = ["Feature"] + feature_names
    covariance_table = [[feature_names[i]] + list(covariance[i]) for i in range(len(feature_names))]
    correlation_table = [[feature_names[i]] + list(correlation[i]) for i in range(len(feature_names))]

    # Print formatted tables
    print("Covariance Matrix:")
    print(tabulate(covariance_table, headers=header, floatfmt=".3f"))

    print("\nCorrelation Matrix:")
    print(tabulate(correlation_table, headers=header, floatfmt=".3f"))

# Example usage
data = np.random.rand(100, 10)
low_level_indices = [0, 1, 2, 3]
high_level_indices = [4, 5, 6, 7, 8, 9]
covariance_all, correlation_all, covariance_low, correlation_low, covariance_high, correlation_high = compute_matrices(data, low_level_indices, high_level_indices)

feature_names_all = [f'Feature {i}' for i in range(data.shape[1])]
feature_names_low = [f'Low Level {i}' for i in low_level_indices]
feature_names_high = [f'High Level {i}' for i in high_level_indices]

create_table(covariance_all, correlation_all, feature_names_all)
create_table(covariance_low, correlation_low, feature_names_low)
create_table(covariance_high, correlation_high, feature_names_high)

```

Plot the significance ($\sigma_{S'}$) for each observable as function of $x_c$ for each scenario, where

```python
import numpy as np
import matplotlib.pyplot as plt

# Example signal and background data
signal_data = np.random.normal(5, 1, 1000)
background_data = np.random.normal(3, 1, 1000)

# Plotting distributions
plt.figure(figsize=(10, 6))
plt.hist(signal_data, bins=30, alpha=0.7, label='Signal', color='blue', density=True)
plt.hist(background_data, bins=30, alpha=0.7, label='Background', color='red', density=True)
plt.xlabel('Observable x')
plt.ylabel('Density')
plt.title('Signal vs Background Distributions')
plt.legend()
plt.show()
```

```python
def compute_tpr_fpr(signal_data, background_data, x_c_values):
    tpr = []
    fpr = []
    for x_c in x_c_values:
        true_positives = np.sum(signal_data > x_c)
        false_positives = np.sum(background_data > x_c)

        tpr.append(true_positives / len(signal_data))
        fpr.append(false_positives / len(background_data))

    return np.array(tpr), np.array(fpr)

x_c_values = np.linspace(0, 10, 100)
tpr, fpr = compute_tpr_fpr(signal_data, background_data, x_c_values)

# Plotting TPR and FPR
plt.figure(figsize=(10, 6))
plt.plot(x_c_values, tpr, label='TPR (Signal Efficiency)', color='blue')
plt.plot(x_c_values, fpr, label='FPR (Background Efficiency)', color='red')
plt.xlabel('Cut Value $x_c$')
plt.ylabel('Efficiency')
plt.title('TPR and FPR as Function of $x_c$')
plt.legend()
plt.grid()
plt.show()
```

```python
def compute_significance(tpr, fpr, N_s, N_b):
    N_s_prime = tpr * N_s
    N_b_prime = fpr * N_b
    significance = N_s_prime / np.sqrt(N_s_prime + N_b_prime)
    return significance

scenarios = [
    (10, 100),
    (100, 1000),
    (1000, 10000),
    (10000, 100000),
]

plt.figure(figsize=(12, 8))

for N_s, N_b in scenarios:
    significance = compute_significance(tpr, fpr, N_s, N_b)
    plt.plot(x_c_values, significance, label=f'N_S={N_s}, N_B={N_b}')

plt.xlabel('Cut Value $x_c$')
plt.ylabel('Significance $\sigma_S\'$')
plt.title('Significance as Function of $x_c$ for Different Scenarios')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd

def calculate_cut_flow_table(data, signal_data, background_data, scenarios):
    cut_flow_tables = {}

    for scenario_name, scenario in scenarios.items():
        N_S = scenario['N_S']
        N_B = scenario['N_B']
        cuts = scenario['cuts']

        cut_flow_data = []
        remaining_signal = len(signal_data)
        remaining_background = len(background_data)

        for observable, x_c in cuts.items():
            true_positives = np.sum(signal_data > x_c)
            false_positives = np.sum(background_data > x_c)

            epsilon_S = true_positives / N_S
            epsilon_B = false_positives / N_B

            N_S_prime = epsilon_S * N_S
            N_B_prime = epsilon_B * N_B

            significance = N_S_prime / np.sqrt(N_S_prime + N_B_prime)

            cut_flow_data.append([observable, epsilon_S, epsilon_B, N_S_prime, N_B_prime, significance])

            # Update remaining counts
            remaining_signal = true_positives
            remaining_background = len(background_data) - false_positives

        cut_flow_tables[scenario_name] = pd.DataFrame(cut_flow_data, columns=['Observable', 'ε_S', 'ε_B', 'N\'_S', 'N\'_B', 'σ_S\''])

    return cut_flow_tables

# Calculate cut flow tables
cut_flow_tables = calculate_cut_flow_table(data, signal_data, background_data, scenarios)

# Display the cut-flow table for each scenario
for name, table in cut_flow_tables.items():
    print(f"\nCut-Flow Table for {name}:\n")
    print(table)
```

### Exercise 8.1

Using numpy, compute the between-class $\bf{S}_B$ and within-class $\bf{S}_W$ covariance matrices defined as:

$$
\bf{S}_B = (\bf{m_2}-\bf{m_1})(\bf{m_2}-\bf{m_1})^T \\
$$
$$
\bf{S}_W = \sum_{i=1,2} \sum_{n=1}^{l_i} (\bf{x}_n^i - \bf{m}_i) (\bf{x}_n^i - \bf{m}_i)^T
$$

where $\bf{m_i}$ are the vectors containing the means for category 1 and 2, here defined as signal and background. Here $\bf{x}_n^i$ is the vector containing the observables for the $n$th example event in category $i$.

### Exercise 8.1

Compute the linear coefficients $\bf{w} = \bf{S_W}^{-1}(\bf{m_2}-\bf{m_1})$. Compare the histogram of the distribution of $F_n^i=\bf{w}^T\bf{x}_n^i$ for the two categories.

What is the maximal significance you can obtain in the scenarios in exercise 5?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Simulated example data (replace with your actual data)
signal_data = np.random.normal(5, 1, (1000, 3))  # 1000 signal events with 3 observables
background_data = np.random.normal(3, 1, (1000, 3))  # 1000 background events with 3 observables

# Compute means for signal and background
mean_signal = np.mean(signal_data, axis=0)
mean_background = np.mean(background_data, axis=0)

# Compute the Between-Class Covariance Matrix
S_B = np.outer(mean_signal - mean_background, mean_signal - mean_background)

# Compute the Within-Class Covariance Matrix
S_W = np.zeros((signal_data.shape[1], signal_data.shape[1]))

for x in signal_data:
    S_W += np.outer(x - mean_signal, x - mean_signal)

for x in background_data:
    S_W += np.outer(x - mean_background, x - mean_background)

# Ensure S_W is invertible
S_W_inv = np.linalg.inv(S_W)

# Print S_B and S_W
print("Between-Class Covariance Matrix S_B:\n", S_B)
print("Within-Class Covariance Matrix S_W:\n", S_W)
```

```python
# Calculate linear coefficients w
w = S_W_inv @ (mean_signal - mean_background)
print("Linear Coefficients w:\n", w)
```

```python
# Compute projections for signal and background
F_signal = signal_data @ w
F_background = background_data @ w

# Combine for ROC calculation
y_true = np.concatenate([np.ones(len(F_signal)), np.zeros(len(F_background))])
y_scores = np.concatenate([F_signal, F_background])
# Compute ROC curve
fpr, tpr, _ = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for F_n')
plt.legend()
plt.grid()
plt.show()
```

```python
def calculate_max_significance(y_true, y_scores, N_S, N_B):
    fpr, tpr, _ = roc_curve(y_true, y_scores)

    max_significance = 0
    for threshold in range(len(tpr)):
        N_S_prime = tpr[threshold] * N_S
        N_B_prime = fpr[threshold] * N_B
        significance = N_S_prime / np.sqrt(N_S_prime + N_B_prime)
        max_significance = max(max_significance, significance)

    return max_significance

# Example scenarios (N_S, N_B)
scenarios = [
    (10, 100),
    (100, 1000),
    (1000, 10000),
    (10000, 100000),
]

# Calculate and print maximum significance for each scenario
for N_S, N_B in scenarios:
    max_sig = calculate_max_significance(y_true, y_scores, N_S, N_B)
    print(f'Max Significance for N_S={N_S}, N_B={N_B}: {max_sig:.2f}')
```
